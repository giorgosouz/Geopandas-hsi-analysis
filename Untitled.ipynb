```python
#insert shapefiles

aerodromia = gpd.read_file("Aerodromia/AERODROMIA.shp",encoding='windows-1253')
kpe = gpd.read_file("kpeshapefile/kpe.shp",encoding='windows-1253')
poleis = gpd.read_file("poleis/poleis.shp",encoding='windows-1253')
periphereies = gpd.read_file("periphereies/periphereies.shp",encoding='windows-1253')
nomoi = gpd.read_file("nomoi_okxe/nomoi_okxe.shp",encoding='windows-1253')
sidhrodromiko_diktyo=gpd.read_file('sidhrodromiko_diktyo/sidhrodromiko_diktyo.shp',encoding='windows-1253')
```

```python
'''line, distance between athens and maleme ariport'''


athens=poleis.loc[poleis['NAME']=='Athens','geometry'].squeeze()
maleme_airport=aerodromia.loc[aerodromia['onoma']=='Μάλεμε','geometry'].squeeze()

line = LineString([athens, maleme_airport])

nom=cascaded_union(nomoi['geometry'])


gpd.GeoSeries([ nom,athens, maleme_airport, line]).plot(cmap='tab10')

print('Total distance is',line.length,'meters')

```

```python
'''Railways within Athens'''
ath=nomoi.loc[nomoi['EDRA']=='ΑΘΗΝΑ','geometry'].squeeze()
diktio=sidhrodromiko_diktyo[sidhrodromiko_diktyo['geometry'].within(ath)]['geometry'].squeeze()
diktio=cascaded_union(diktio)

gpd.GeoSeries( [ath,diktio]).plot(cmap='tab10')

"""total railway length"""

diktyo=cascaded_union(sidhrodromiko_diktyo['geometry'])
print('Total length of  railway is',diktyo.length,'meters')
gpd.GeoSeries([ nom, diktyo]).plot(cmap='tab10')


```

```python
'''number of airports in Crete'''
periph=periphereies.loc[periphereies['PER']=='Π. ΚΡΗΤΗΣ','geometry'].squeeze()
airs=aerodromia[aerodromia['geometry'].within(periph)]['geometry']
print("The number of airports in Crete is",len(airs))
airs=cascaded_union(airs)

print(gpd.GeoSeries( [periph,airs]).plot(cmap='Paired'))
```

```python
"""find closest kpe in Athens"""

arg_min=(kpe.distance(athens)).idxmin()
kpe.iloc[arg_min]
```

```python
"""area of interest on kpe(buffer 20000)"""

kk=kpe.buffer(20000).squeeze()
kk=cascaded_union(kk)
gpd.GeoSeries([ nom, kk]).plot(cmap='tab10')
```

```python

"""thermals of Kastoria"""

thermals=pd.read_csv('timeseries.csv')
thermals.head()

#take the rows from 2000

thermals_2000=thermals.loc[thermals['YEAR']==2000]
thermals=thermals_2000[['T2M','DOY']]
plt.plot(thermals['DOY'],thermals['T2M'])
plt.xlabel('DOY')
plt.title(" Temperature over year 2000, Kastoria")
plt.ylabel('Temperature ')
plt.show()
```

```python
#summer mean thermals for every year


thermals=pd.read_csv('timeseries.csv')
thermals_summer=thermals.loc[thermals['DOY'].between(150,240)]
thermals_summer
summer=thermals_summer.groupby('YEAR').mean()
summer.head()
```

```python
data=pd.read_csv('eurostat/nama_10_gdp_1_Data.csv')
translator=pd.read_csv('eurostat/UID_ISO_FIPS_LookUp_Table.csv')

# dind unique values- replace odd names
xwres=np.unique(data['GEO'])
data.replace('Germany (until 1990 former territory of the FRG)','Germany',inplace=True)
data.replace('Kosovo (under United Nations Security Council Resolution 1244/99)','Kosovo',inplace=True)
xwres=np.unique(data['GEO'])

# make new columns
data['LAT']=np.arange(len(data))
data['LON']=np.arange(len(data))


# extract lat long
for i in xwres:
    a=translator.loc[translator['Combined_Key']==i][['Lat','Long_']]
    data.loc[data['GEO']==i,'LAT']=a.iloc[0,0]
    data.loc[data['GEO']==i,'LON']=a.iloc[0,1]


    
#preproccess
data.TIME=pd.to_datetime(data.TIME, format='%Y')
np.unique(data.Value)
data.replace(',','',regex=True,inplace=True)
data['Value'] = data['Value'].apply(pd.to_numeric,errors='coerce')
data.Value.replace(':',data.Value.mean(),inplace=True)
data.fillna(round(data.Value.mean(),2),inplace=True)

data.to_csv('tax.csv')
data
```
